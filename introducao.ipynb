```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 6, 150) # esse mtodo faz com que possa ser gerado um array entre tempo final e inicial. Exemplo: de 0 segundos a 6 segundos com 100 amostras
y = np.cos(t) #funao cosseno na biblioteca numpy

plt.figure(1) #figure  a "figura"geral do grfico, a estrutura que receber os grficos e eixos
plt.plot(t, y)
plt.title("Grfico do Cosseno") # adiciona ttulo para o grfico
plt.xlabel("Tempo em segundos") #adiciona legenda para o eixo X
plt.ylabel("Amplitude") #adiciona legenda para o eixo Y
plt.show() #apresenta o grfico

y1 = np.sin(t)
plt.figure(2) 
plt.plot(t, y1)
plt.title("Grfico do Seno") 
plt.xlabel("Tempo em segundos") 
plt.ylabel("Amplitude") 
plt.show() 
```
