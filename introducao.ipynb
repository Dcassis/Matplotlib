```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 6, 150) # esse método faz com que possa ser gerado um array entre tempo final e inicial. Exemplo: de 0 segundos a 6 segundos com 100 amostras
y = np.cos(t) #funçao cosseno na biblioteca numpy

plt.figure(1) #figure é a "figura"geral do gráfico, a estrutura que receber os gráficos e eixos
plt.plot(t, y)
plt.title("Gráfico do Cosseno") # adiciona título para o gráfico
plt.xlabel("Tempo em segundos") #adiciona legenda para o eixo X
plt.ylabel("Amplitude") #adiciona legenda para o eixo Y
plt.show() #apresenta o gráfico

y1 = np.sin(t)
plt.figure(2) 
plt.plot(t, y1)
plt.title("Gráfico do Seno") 
plt.xlabel("Tempo em segundos") 
plt.ylabel("Amplitude") 
plt.show() 
```
