Subplot e Subplots - figura com vários gráficos

```python
# Como plotar vários gráficos dentro da mesma figura?

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 500)
c = np.cos(x)
s = np.sin(x)
plt.figure("Gráficos", figsize=(8,4))
# para ajustar as dimensões dos gráficos use SUBPLOTS_AJUST
plt.subplots_adjust(
    left=0.13,
    right=0.91,
    top=0.9,
    bottom=0.14,
    wspace=0.35,
    hspace=0.4
)

plt.subplot(1,2,1) # significa (numero de linhas, colunas e posiçao do gráfico) uma linha e 2 colunas, o último 1 significa a posiçao desse gráfico - primeira coluna.
plt.plot(x, c)
plt.title("Gráfico do Cosseno")
plt.xlabel("Eixo de Tempo")
plt.ylabel("Eixo de Amplitude")
plt.grid() # para adicionar as grids

plt.subplot(1,2,2)
plt.plot(x, s)
plt.title("Gráfico do Seno")
plt.xlabel("Eixo de Tempo")
plt.ylabel("Eixo de Amplitude")
plt.grid()

################################### SUBPLOTS - subplot agrupado com um objeto ####################################

x = np.arange(0, 5, 0.1)
y = x**2
y2 =x**3
y3 =x**4
y4 =x**5


fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(8,6))
plt.suptitle("Gráficos com Subplots")
plt.subplots_adjust(
    left=0.13,
    right=0.91,
    top=0.9,
    bottom=0.14,
    wspace=0.35,
    hspace=0.6
)

ax1.plot(x,y)
ax1.set_title("Função de Segundo Grau $xˆ2$")
ax1.set_xlabel("Tempo")
ax1.set_ylabel("Amplitde")

ax2.plot(x, y2)
ax2.set_title("Função de Terceiro Grau $xˆ3$")
ax2.set_xlabel("Tempo")
ax2.set_ylabel("Amplitde")

ax3.plot(x, y3)
ax3.set_title("Função de Quarto Grau $xˆ4$")
ax3.set_xlabel("Tempo")
ax3.set_ylabel("Amplitde")

ax4.plot(x, y4)
ax4.set_title("Função de Quinto Grau $xˆ5$")
ax4.set_xlabel("Tempo")
ax4.set_ylabel("Amplitde")
```
