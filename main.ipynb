```python
# Como plotar 2 gráficos dentro da mesma figura?

import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0, 100) # o Arange importa valores inteior s de 0 a 100
y = np.cos(t)
y1 = np.sin(t)

plt.figure("Gráfico", figsize=(6,4))
plt.plot(t,y)
plt.plot(t, y1)
plt.show()

t = np.linspace(0, 2*np.pi, 100) 
y = np.cos(t)
y1 = np.sin(t)

plt.figure("Gráfico 2", figsize=(6,4))
plt.plot(t,y)
plt.plot(t, y1)
plt.title("Gráfico de Seno e Cosseno")
plt.xlabel("Eixo de tempo")
plt.ylabel("Eixo da amplitude")
plt.grid() # grid adiciona uma grid ao gráfico
plt.show()
```
